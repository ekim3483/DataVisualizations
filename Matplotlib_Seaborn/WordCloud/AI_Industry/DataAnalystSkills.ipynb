```python
import numpy as np
import pandas as pd

%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import matplotlib.patches as mpatches
from PIL import Image

!pip install seaborn wordcloud 

import seaborn as sns

import wordcloud

mpl.style.use('dark_background')
```

```python
df = pd.read_csv('ai_job_market.csv')
```

```python
df_DA=df[(df['job_title']=='Data Analyst')]
```

```python
df_DA.head()
```

```python
df_DA_skillset=pd.DataFrame(df_DA['skills_required'])
```

```python
df_DA_skillset.head()
```

```python
df_DA_skillset = df_DA_skillset['skills_required'].str.split(',',expand=True)
```

```python
df_DA_skillset.head(3)
```

```python
df_DA_skillset.to_csv('df_DA.txt', sep='\t', header=False, index=False)
```

```python
df_DA = open ("df_DA.txt").read()
```

```python
from wordcloud import WordCloud
df_DA_skills = WordCloud()
df_DA_skills.generate(df_DA)
```

```python
fig = plt.figure(figsize=(14, 18))

# display the cloud
plt.imshow(df_DA_skills, interpolation='bilinear')
plt.axis('off')
plt.show()
```
